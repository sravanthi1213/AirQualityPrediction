```python
import numpy as np

# Confusion matrix data
svm = np.array([[788, 62, 8, 34],
                [241, 490, 7, 50],
                [220, 76, 481, 61],
                [136, 37, 15, 516]])

# True positives, false positives, false negatives, true negatives
TP = np.diag(svm)
FP = np.sum(svm, axis=0) - TP
FN = np.sum(svm, axis=1) - TP
TN = np.sum(svm) - (TP + FP + FN)

# Accuracy
accuracy = np.sum(TP) / np.sum(svm)

# Precision
precision = TP / (TP + FP)

# Recall
recall = TP / (TP + FN)

# F1-score
f1_score = 2 * (precision * recall) / (precision + recall)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1_score)
```

```python
import numpy as np

# Confusion matrices for each model
confusion_matrices = {
    "Logistic Regression": logistic_regression,
    "Naive Bayes": naive_bayes,
    "SVM": svm,
    "XGBoost": xgboost,
    "Decision Tree": decision_tree
}

# Metrics storage
metrics = {}

for model, matrix in confusion_matrices.items():
    # True positives, false positives, false negatives, true negatives
    TP = np.diag(matrix)
    FP = np.sum(matrix, axis=0) - TP
    FN = np.sum(matrix, axis=1) - TP
    TN = np.sum(matrix) - (TP + FP + FN)

    # Accuracy
    accuracy = np.sum(TP) / np.sum(matrix)

    # Precision
    precision = np.mean(TP / (TP + FP))

    # Recall
    recall = np.mean(TP / (TP + FN))

    # F1-score
    f1_score = np.mean(2 * (precision * recall) / (precision + recall))

    # Storing metrics
    metrics[model] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-score": f1_score
    }

# Output metrics
for model, metric_values in metrics.items():
    print(f"{model} Metrics:")
    print("Accuracy:", metric_values["Accuracy"])
    print("Precision:", metric_values["Precision"])
    print("Recall:", metric_values["Recall"])
    print("F1-score:", metric_values["F1-score"])
    print()
```
